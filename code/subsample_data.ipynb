```python
# training/test split - use 70/30 as in paper

n = len(asample2)
train_idx = np.random.choice(asample2.index, size=int(0.7 * n), replace=False)
test_idx = asample2.index.difference(train_idx)

X = asample2.drop('Cm_ltotalincome', axis=1)
y = asample2['Cm_ltotalincome']

X_train = X.loc[train_idx]
X_test = X.loc[test_idx]

y_train = y.loc[train_idx]
y_test = y.loc[test_idx]
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error

model = lgb.LGBMRegressor(
    n_estimators = 100,
    max_depth = 3,
    learning_rate=0.1
)

model.fit(X_train, y_train)
preds = model.predict(X_test)
mse = mean_squared_error(y_test, preds)
```
